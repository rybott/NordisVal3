# With OCR

```python
from docling.document_converter import DocumentConverter, PdfFormatOption
from docling.datamodel.base_models import InputFormat
from docling.datamodel.pipeline_options import PdfPipelineOptions, EasyOcrOptions, TesseractCliOcrOptions

source = "/home/coder/Repos/Tax_Processing/Automating/Data/PDFs/2017 S Corp/111 S Corp Tax Return IS no Statements.pdf"
engine = "easyocr"   # or: "tesseract"

ocr_opts = EasyOcrOptions(lang=["en"]) if engine.lower()=="easyocr" else TesseractCliOcrOptions(lang=["eng"])

pipe = PdfPipelineOptions(do_ocr=True, ocr_options=ocr_opts)
conv = DocumentConverter(format_options={InputFormat.PDF: PdfFormatOption(pipeline_options=pipe)})

doc = conv.convert(source).document
print(doc.export_to_markdown())
print("test")
```
